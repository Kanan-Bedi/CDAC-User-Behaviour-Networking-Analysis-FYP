```python
def chunk_iterator(file_path, chunk_size=100):
    """Reads the file in chunks (if large) to handle memory efficiently."""
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        yield chunk
```

```python
def merge_all_content():
    """Merges the pre-processed files and saves the merged result."""
    print("Merging pre-processed content...")

    # Initialize with the first file's content
    merged_df = pd.concat(chunk_iterator(output_dir / '/kaggle/input/preprocessed-cert/email_preprocessed.csv'))

    # Append content from the other files
    merged_df = pd.concat([merged_df, *chunk_iterator(output_dir / '/kaggle/input/preprocessed-cert/file_preprocessed.csv')])
    merged_df = pd.concat([merged_df, *chunk_iterator(output_dir / '/kaggle/input/preprocessed-cert/http_preprocessed.csv')])

    # Save the merged content to a file
    merged_df.to_pickle(output_dir / 'merged_content.pkl')
    print(f"Merged content saved to: {output_dir / 'merged_content.pkl'}")
```

```python
from pathlib import Path

# Define the output directory path
output_dir = Path('./_output')

# Create the directory if it doesn't exist
output_dir.mkdir(parents=True, exist_ok=True)

print(f"Directory '{output_dir}' created successfully (or already exists).")
```
